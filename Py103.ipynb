# Object identity vs Equality

```python
class Dog:
    pass

# Create two separate Dog objects
dog1 = Dog()
dog2 = Dog()

# Give both dogs the same name value
dog1.name = "Buddy"
dog2.name = "Buddy"

print("dog1 name:", dog1.name)
print("dog2 name:", dog2.name)

# Identity check: are these two variables pointing to the SAME object?
print("dog1 is dog2:", dog1 is dog2)

# Equality check: do these two objects have the SAME VALUE?
# For our simple Dog class, Python doesn't know how to compare values yet,
# so this will still be based on identity.
print("dog1 == dog2:", dog1 == dog2)
```

# Methods & the Role of Self

```python
class Dog:
    def speak(self):
        print("Woof!")
# Here, def speak(self): is the method definition.
```

When you DEFINE the method

You’re writing the recipe for what a dog does when it speaks:

Here you must write self because you’re saying:

“Dear Python, when some dog speaks, I need to know which dog it is.”

So self is like an empty seat reserved for “this dog”.

When you CALL the method

Now you actually have a dog:

Here, you already said which dog: it’s buddy (the thing on the left of the dot).

Because of that, Python does the boring part for you:
You write: buddy.speak()

Python turns it into: Dog.speak(buddy) behind the scenes

Inside speak, self now means buddy

* We write self in the definition so the method has a way to receive the object.

* We do not write self in the call because the object before the dot is already being passed in.

```python
class Dog:
    def speak(self):
        print("Woof!")

buddy = Dog()
luna = Dog()

buddy.speak() # here self is buddy
luna.speak() # here self is luna

# So self changes depending on which object is on the left of the dot.
# It’s always “this exact object that is speaking right now.”
```

# Keeping Objects Valid

```python
class Pet:
    def __init__(self, name, age):
        self.name = name
        self.age = age
Pet("", -5)
```

We want to add guard-style checks inside __init__ so the object ends up with safe values.

```python
class Pet:
    def __init__(self, name, age):
        if not name.strip():      # name is "" or "   "
            self.name = "Unknown"
        else:
            self.name = name

        self.age = age            # we’ll validate age later
```

if not name.strip() → “is the name empty or just spaces?” (works for “blank or whitespace name”)

We’re using .strip() as a test, not as the final stored value (at least in this version)



```python
class Pet:
    def __init__(self, name, age):
        cleaned_name = name.strip()

        if not cleaned_name:           # empty after stripping?
            self.name = "Unknown"
        else:
            self.name = cleaned_name

        self.age = age                 # age still unvalidated for now
```

```python
class Pet:
    def __init__(self, name, age):
        cleaned_name = name.strip()

        if not cleaned_name:
            self.name = "Unknown"
        else:
            self.name = cleaned_name

        self.age = age

pet1 = Pet("   ", 3)
pet2 = Pet("Max", 5)

print(pet1.name, pet1.age)
print(pet2.name, pet2.age)

```

```python
class Pet:
    def __init__(self, name, age):
        cleaned_name = name.strip()

        if not cleaned_name:
            self.name = "Unknown"
        else:
            self.name = cleaned_name

        if age < 0:
            self.age = 0
        else:
            self.age = age

pet1 = Pet("   ", 3)
pet2 = Pet("Max", 5)
pet3 = Pet("Bella", -2)

print(pet1.name, pet1.age)
print(pet2.name, pet2.age)
print(pet3.name, pet3.age)

```

# Building Objects from Input

**Object**

An object is like a custom data box in memory that has named pieces of data inside it.
Example: a Guest object might have a name and a age.

**Class**

A class is a blueprint for making objects.
Example: the Guest class is the recipe; each Guest(...) you create is one actual guest object.

**Constructor** (__init__)

The constructor is a special function inside a class that runs when you make a new object.
Example:

class Guest:

    def __init__(self, name):
        self.name = name

When you call Guest("Sam"), Python runs __init__ and stores "Sam" in self.name.

**List**

A list is an ordered collection of items in Python, written with [].
Example: guests = [] starts an empty list; guests.append(one_guest) adds an item.

**Loop**

A loop repeats code several times.
We’ll use it to repeat: ask user → build object → add to list.

**Sentinel question**

A sentinel question is a yes/no question we ask to decide when to stop the loop.
Example: "Add another guest? (y/n): " — if user says n, we stop.

```python
# class section
class Guest:                     # define a blueprint called Guest
    def __init__(self, name):    # constructor: runs when we make a Guest
        self.name = name         # store the given name inside the object

# runtime section
name = input("Enter guest name: ")  # ask the user for a name, save it in variable 'name'
guest1 = Guest(name)                # create a Guest object using that name and store it in 'guest1'
```

```python
# class section
class Guest:
    def __init__(self, name):
        self.name = name

# runtime section
guests = []                         # empty list to store Guest objects

while True:                         # repeat until we break
    name = input("Enter guest name: ")   # 1) ask for a name
    guest = Guest(name)                  # 2) build a Guest object
    guests.append(guest)                 # 3) store it in the list

    more = input("Add another guest? (y/n): ")  # sentinel question
    if more == "n":                             # if user says no
        break                                   # stop the loop
```

For our pet: “happiness is always between 0 and 100.”

```python
class Pet:
    def __init__(self, name, happiness):
        self.name = name

        # clamp in __init__
        if happiness < 0:
            happiness = 0   #sloppy pattern: USE SELF...
        if happiness > 100:
            happiness = 100

        self.happiness = happiness

    def play(self):
        # change state
        self.happiness = self.happiness + 30

        # clamp after change
        if self.happiness < 0:
            self.happiness = 0
        if self.happiness > 100:
            self.happiness = 100

pet = Pet("Fifi", 80)
pet.play()   # +30
pet.play()   # +30
pet.play()   # +30

print(pet.happiness)
```

# Introducing Inheritance

We put all the shared stuff (like name, hunger, feed()) into Pet, so other classes (like Dog, Cat) don’t have to repeat that code.

# Polymorphism Basics: one action, different behaviors

Remember the pet show: judge says “Speak!” to every animal: (Some code is calling the method)

* Dog barks
* Cat meows
* Bird chirps

Same request, different result. That’s polymorphism in real life.


```python
class Pet:
    def speak(self):
        print("Some generic pet sound")

class Dog(Pet):
    def speak(self):
        print("Woof!")

class Cat(Pet):
    def speak(self):
        print("Meow!")
```

```python
class Pet:
    def speak(self):
        print("Some generic pet sound")

class Dog(Pet):
    def speak(self):
        print("Woof!")

class Cat(Pet):
    def speak(self):
        print("Meow!")

class Bird(Pet):
    def speak(self):
        print("Chirp!")

pets = [Dog(), Cat(), Bird()]

for pet in pets:
    pet.speak()
```

“Which speak() runs?” is decided by the object’s class (Dog vs Cat), not by the loop code.
So the call stays the same, the object changes, and Python picks a different speak() each time. That’s polymorphism in action.
So the OOP feature is inheritance + method overriding. That setup is what makes the polymorphic loop line pet.speak() possible.

* Pet defines speak
* Dog, Cat, Bird inherit from Pet
* Each one overrides speak with its own behavior

```python
# Base class for all animals
class Animal:
    def __init__(self, name, age):
        # Internal (encapsulated) attributes
        self._name = name
        self._age = age

    # Getter for name
    def get_name(self):
        return self._name

    # Setter for name with simple validation
    def set_name(self, name):
        if isinstance(name, str) and name.strip() != "":
            self._name = name
        else:
            print("Name must be a non-empty string")

    # Getter for age
    def get_age(self):
        return self._age

    # Setter for age with simple validation
    def set_age(self, age):
        if age >= 0:
            self._age = age
        else:
            print("Age cannot be negative")

    # Default behavior for all animals
    def speak(self):
        print("Some generic animal sound")

    # Human-readable string for printing Animal objects
    def __str__(self):
        return f"Animal(name={self._name}, age={self._age})"


# Dog is-a Animal (inheritance)
class Dog(Animal):
    def __init__(self, name, age, breed):
        # Call parent (Animal) __init__ to set name and age
        super().__init__(name, age)
        # Dog-specific internal attribute
        self._breed = breed

    # Dog-specific version of speak (overriding)
    def speak(self):
        print(f"Woof! My name is {self._name}")

    # Human-readable string for printing Dog objects
    def __str__(self):
        return f"Dog(name={self._name}, age={self._age}, breed={self._breed})"


# --- Test code below ---

# Create one Animal and one Dog
a = Animal("Buddy", 5)
d = Dog("Max", 3, "Beagle")

# Show overriding of speak()
a.speak()  # uses Animal.speak
d.speak()  # uses Dog.speak

# Show __str__ overriding
print(a)   # uses Animal.__str__
print(d)   # uses Dog.__str__

# More animals for polymorphism demo
a2 = Animal("Luna", 2)
d2 = Dog("Charlie", 4, "Poodle")

# List with mixed Animal and Dog objects
zoo = [a, d, a2, d2]

# Polymorphism: same call, different behavior
for animal in zoo:
    animal.speak()
```

When you design a class with methods like deposit(), withdraw(), describe(), you’re doing abstraction:

Other code doesn’t know or care how it works inside, just what it can do.
__ str__ is just a special method Python calls when you print(obj) or do str(obj).

It doesn’t hide logic.
It just controls the string representation.

So:

You can have abstraction without __ str__ (just methods and hidden internals).
You can have __ str__ without good abstraction (a messy class that only prints nicely).

__ str__ = “How should I look as text?”

Abstraction = “What do I show vs hide about how I work?”
